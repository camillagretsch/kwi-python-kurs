Führen Sie als erstes wieder den unten stehenden Code aus, damit Sie wieder die Turtle Befehle ausführen können.

````python
import ipyturtle3 as turtle

def prepareTurtleGraphics(width=500, height=250):
    myCanvas = turtle.Canvas(width=width, height=height)
    display(myCanvas)
    turtleScreen = turtle.TurtleScreen(myCanvas)
    turtleScreen.clear()
    t = turtle.Turtle(turtleScreen)
    t.shape("turtle")
    turtleScreen.bgcolor("floralwhite")
    return (t, turtleScreen)
````

Der unten stehende Code zeichnet ein Sechseck mit Seitenlänge 50.

## 1. Der Zuweisungsoperator =

Wir haben mit a = 50 einen neuen Befehl kennengelernt. Er gehört zur Kategorie “Zuweisung” (eng. assignment). a = 50 ist ein Beispiel für eine (einfache) Zuweisung. Bei einer Zuweisung wird immer ein **Wert in einer Variable gespeichert**. **Links** steht immer der **Variablename**, rechts können komplexere Befehle notiert werden. In unserem Beispiel ist es einfach die Zahl 50.

<img src="./bilder/03_Variablen_Zuweisungsoperator.png" style="height:100px"/>

In Python (und vielen anderen Programmiersprachen auch) ist das Gleichheitszeichen ein Operator. Der Operator wird Zuweisungsoperator genannt und ist nicht mit dem mathematischen Gleichheitszeichen zu verwechseln! Man kann die Zuweisung auch lesen als “a erhält den Wert 275”.

---
**Definition: Wert**: Programme arbeiten mit Informationen in Form von zum Beispiel Zahlen und Text. Ein Wert ist eine konkrete Information. Es gibt zum Beispiel Zahlenwerte (z.B. 25 oder 3.14), Textwerte (z.B. "Hello, Dave!") und Wahrheitswerte (True und False). Dies sind nur Beispiele. Wir werden verschiedene Werte verwenden.

## 2. Wo wird der Wert gespeichert?
Eine Variable ist ein Speicherplatz mit einem Namen. Das Computerbauteil, in dem der Speicherplatz
für eine Variable reserviert wird, heisst Arbeitsspeicher. Der Arbeitsspeicher für Standardnotebooks ist typischerweise zwischen 8GB und
32GB gross. Nur ein Teil davon wird zur Speicherung von Variablen verwendet.

Man kann sich diesen Teil des Speichers wie eine Tabelle mit zwei Spalten vorstellen. In der ersten
Spalte steht der Variablenname, in der zweiten Spalte steht der gespeicherte Wert. Es wird erst
dann eine Zeile der Tabelle hinzugefügt, wenn man das Programm ausführt und eine Zuweisung
verarbeitet wurde.

Schauen wir uns das mal für folgenden Code an:

```Python
1 (t, tScreen) = prepareTurtleGraphics()
2 a = 100
3 t.forward(a)
4 t.left(90)
5 t.forward(a)
6 t.left(90)
7 t.forward(a)
8 t.left(90)
9 t.forward(a)
10 t.left(90)
```

Die Tabelle ist zu Beginn “leer”. Es ist keine Variable vorhanden, es wird kein Wert gespeichert. Nun wird Zeile für Zeile verarbeitet. Erst nachdem Zeile zwei verarbeitet wurde, befindet sich ein neuer Eintrag in der Tabelle. Es wird die Variable mit dem Namen a und der Wert 100 in einer Zeile der Tabelle hinzugefügt.

| Variablename   |      WERT      | 
|:----------|:---------------|
|- | -|
Zu Beginn des Programmes ist die Tabelle leer.

| Variablename   |      WERT      | 
|:----------|:---------------|
|a | 100|
Erst **nachdem** die Zeile 2 abgearbeitet wurde existiert die Variable a und enthält den Wert 100.


Wenn die Variable existiert, kann über ihren Namen auf den Wert zugegriffen werden. 

Dies geschieht in den Zeilen 3, 5, 7 und 9. Wenn Sie vorher auf die Variable zugreifen würden, würden Sie einen **NameError** erhalten, da die Variable noch nicht existiert, das Programm also den Namen `a` noch nicht kennt.

## 3. Die Input-Funktion

Mit der eingebauten Funktion input kann man **während der Programmausführung** Werte eingeben. Dies nennt sich **Benutzereingabe** (eng. user input). Dadurch kann ein Benutzer des Programms den **Programmablauf** beeinflussen. Über die Konsole kann man etwas eintippen. Wenn man etwas in der Konsole eintippt, dann benötigt man für die Eingabe keine Anführungszeichen. *Python behandelt alles, was mit input über die Konsole eingetippt wird als Text!* Der Programmablauf wird solange gestoppt, bis der Benutzer oder die Benutzerin eine Eingabe über die Tastatur durchführt und diese mit der Return-Taste (“Enter”) abschliesst.

Probieren Sie folgenden Code aus um zu sehen, wie die Benutzereingabe mithilfe der input()-Funktion funktioniert.

```Python
vorname = input("Wie lautet ihr Vorname? ")
print("Sie haben", vorname, "eingegeben")
```
⚠️ **Achtung:** Bei print kann man mehrere verschiedene Werte ausgeben, wenn man sie mit Kommas trennt.
