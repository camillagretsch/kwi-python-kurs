## 1. Ausführung eines Programms, einfache Befehle und Kommentare

Wie ein Rezept aus der Beschreibung der einzelnen Arbeitsschritte besteht, ist Code aus Befehlen aufgebaut.

Letzte Woche haben Sie den Befehl `print()` kennengelernt und damit Ihren Namen und Ihre Klasse ausgegeben:

```Pyton

print("3k")
```

Wenn Sie es ausführten, lautete die Ausgabe:

```

```

Wenn Programmcode ausgeführt wird, werden die Befehle von oben nach unten abgearbeitet. 

Das sehen Sie daran, dass die Resultate dieser Befehle in der gleichen Reihenfolge ausgegeben werden, wie die `print()`-Befehle.

Merken Sie sich folgendes:

**Code wird immer von oben nach unten ausgeführt**, also zuerst die erste Zeile der Codezelle, dann die zweite Zeile usw.

### 1.1. Befehle

Befehle sind eigentlich Aufrufe von Funktionen. Einfache Befehle werden immer gleich geschrieben:

```Python
befehlsname(argumente, mit, Kommas, getrennt)
```

Zwischen den Klammern können dem Befehl keines, eines oder mehrere Argumente übergeben werden. Der `print()`-Befehl kann beliebig viele Argumente entgegennehmen:

`print()` gibt nicht mehr aus, als einen Zeilenumbruch,  
`print("hallo")` gibt `Hallo` aus, gefolgt von einem Zeilenumbruch,  
`print("heute schreibe", "ich lustige", "Programme", "heute schreibe ich lustige Programme", "!")` gibt `heute schreibe ich lustige Programme !` aus.

<details>

         ► Wenn Sie mehr dazu wissen möchten, klappen Sie diesen Bereich aus. ◀︎

`t` repräsentiert eine Turtle und `tScreen` das Zeichenfeld.

Weitere Turtles können erstellt werden mit:

```Python
blau = turtle.Turtle(tScreen)
blau.color("blue")
blau.shape("turtle")
```

Optional kann auch die Grösse des Zeichenfelds festgelegt werden. 

Standardwerte: `width=500`, `height=250`

```Python
(t,tScreen) = prepareTurtleGraphics(width, height)
```
</details>

````python
# Führen Sie diese Zelle aus, damit Sie weiter unten Turtlegrafiken machen können...

import ipyturtle3 as turtle

def prepareTurtleGraphics(width=500, height=250):
    myCanvas = turtle.Canvas(width=width, height=height)
    display(myCanvas)
    turtleScreen = turtle.TurtleScreen(myCanvas)
    turtleScreen.clear()
    t = turtle.Turtle(turtleScreen)
    t.shape("turtle")
    turtleScreen.bgcolor("floralwhite")
    return (t, turtleScreen)
````

## 2. Datentypen für Zahlen

Vielleicht ist Ihnen aufgefallen, dass die Argumente der Turtle-Befehle `forward()`, `left()` etc. anders dargestellt werden als die bisher dem `print()`-Befehl übergebenen Argumente:

```Python
print("Hallo!")
t.forward(100)
t.left(45)
```

Dies liegt daran, dass Sie korrekterweise Zahlen und keinen Text übergeben haben.

***Spezialfall `print()`-Befehl***

Der `print()`-Befehl ist etwas speziell. Er gibt alles aus, was er an Argumenten erhält. 

Er kann *beliebig viele* Argumente *aller möglichen Datentypen* entgegennehmen. Beim Turtle-Befehl `forward()` ist das anders.

### 2.1. TypeError: falscher Datentyp!

Der Aufruf

```Python
t.left("45")
```

würde zu einer Fehlermeldung führen, weil Der Turtle-Befehl `left()` eine Zahl und keinen Text erwartet.

In der Regel erwarten Funktionen für die Argumente vorher festgelegte Datentypen. Falls der Typ falsch ist, gibt es einen sogenannten *TypeError*.

## 3. Farbige Kunstwerke erstellen mit der Turtle

Nun kommt Farbe ins Spiel! Die Turtle kann auch die Stiftfarbe und -Dicke verändern und die Figuren, die sie zeichnet, ausmalen. Damit wird es möglich, mit der Turtle farbige Kunstwerke zu erstellen.

| Befehl   |      Abkürzung      |  Effekt |
|:----------|:---------------|:-------|
|color() | | Farbe der Turtle (und des Stifts), z.B. color("red") |
|pencolor() | | Stiftfarbe (bei ausgefüllten Formen: Rand), z.B. pencolor("red") |
|begin_fill()<br/>end_fill() | | Was zwischen `begin_fill()` und `end_fill()` gezeichnet wird,<br/>wird ausgefüllt. |
|fillcolor() | | Füllfarbe|
|pensize() | width() | Stiftbreite |
|circle(r) | circle(r) | Kreis mit Radius r |
|isdown() | | Gibt den Status des Stifts zurück: `true` für *down*, `false` für *up*. |

Für die **Farben** können Sie gängige Farbnamen verwenden, z.B. "red", "green","blue", "orange", "purple", ... [Hier finden Sie 140 Farben mit ihren Codes.](https://trinket.io/docs/colors)<!--[Hier finden Sie eine Liste der möglichen Farbnamen](https://www.w3schools.com/colors/colors_names.asp)-->

Auch die Hintergrundfarbe des Zeichenbereichs können Sie verändern. Den folgenden Befehl wenden Sie nicht auf die Turtle `t` an, sondern auf die Zeichenfläche `tScreen`:

| Befehl    | Effekt |
|:----------|:-------|
| bgcolor() | Hintergrundfarbe definieren<br/>z.B. tScreen.bgcolor("lavender") |

#### 1.3.3. Weitere Turtlebefehle

Die Turtle kennt noch weitere Befehle. 

Sie kann beispielsweise den Stift hochheben und wieder absetzen und sie kann sich auch verstecken und wieder zeigen.

| Befehl   |      Abkürzung      |  Effekt |
|:----------|:---------------|:-------|
|penup() | pu() | hebt den Stift von der Zeichenfläche hoch, damit die Turtle nicht mehr zeichnet. |
|pendown() | pd() | setzt den Stift wieder auf die Zeichenfläche, damit die Turtle wieder zeichnet. |
|hideturtle() | ht() | macht die Turtle unsichtbar.|
|showturtle() | st() | macht die Turtle sichtbar. |

| Befehl   |      Abkürzung      |  Effekt |
