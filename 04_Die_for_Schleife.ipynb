Eine for-Schleife besteht jeweils aus dem **Schleifenkopf**, welcher die **Schleifenvariable** `i` enthält und dem **Schleifenkörper**.
Der **Schleifenkopf** bestimmt dabei die Anzahl Durchläufe und die **Schleifenvariable** und sieht so aus:

```Python
for i in range(anzahl_wiederholungen):
```
Der **Schleifenkörper** enthält den Code, der mehrfach ausgeführt werden soll. Er ist durch **Einrücken** markiert. 

Für das obige Beispiel sieht die gesamte for-Schleife so aus:

```Python
for i in range(360): # Schleifenkopf (Achten Sie auch auf den Doppelpunkt)
    t.fd(3)          # Schleifenkörper (eingerückter Block)
    t.lt(1)
```

**Aufgabe 2**. Wofür brauchen wir eine Schleifenvariable überhaupt?
Versuchen Sie mal herauszufinden, was mit der Schleifenvariable passiert während dem ausführen einer for-Schleife. Benutzen sie dazu z.B. einen `print()`-Befehl, der jeweils die Schleifenvariable ausgibt.

````python
# Testen Sie hier, was mit der Schleifenvariable passiert.

for i in range(5):
    print(i)
````

Sie sollten bemerkt haben, dass die Schleifenvariable mit dem Wert 0 startet und dann hochzählt bis 1 unter den Wert, welcher in der range-Funktion steht. Die Schleifenvariable existiert nur solange die Schleife ausgeführt wird. Nach dem letzten Durchgang wird nochmals der Schleifenkopf ausgeführt und die Schleifenvariable wird wieder gelöscht.
**Wichtig** ist, dass der Computer beim Ausführen des Codes vor jedem Schleifendurchgang auch den Schleifenkopf ausführt und dabei die Schleifenvariable erhöht. Bei folgendem Code würde dies folgendes bedeuten:

```Python
1 for i in range(4):
2    t.fd(100)
3    t.lt(90)
4 t.fd(100)
```


| Zeile   |      Wert der Schleifenvariable      |  Kommentar |
|:----------|:---------------|:-------|
|1 | i = 0 | Eine Variable **i mit dem Wert 0** wird kreiert, nachdem die Zeile ausgeführt wurde. |
|2 | i = 0 | i speichert immer noch den Wert 0. |
|3 | i = 0 | i speichert immer noch den Wert 0. |
|1 | i = 1 | **i wird auf 1** gesetzt, nachdem die Zeile ausgeführt wurde. |
|2 | i = 1 | i speichert immer noch den Wert 1. |
|3 | i = 1 | i speichert immer noch den Wert 1. |
|1 | i = 2 | **i wird auf 2** gesetzt, nachdem die Zeile ausgeführt wurde. |
|2 | i = 2 | i speichert immer noch den Wert 2. |
|3 | i = 2 | i speichert immer noch den Wert 2. |
|1 | i = 3 | **i wird auf 3** gesetzt, nachdem die Zeile ausgeführt wurde. |
|2 | i = 3 | i speichert immer noch den Wert 3. |
|3 | i = 3 | i speichert immer noch den Wert 3. 
|**1**|**-**| **nach dem Ausführen des Schleifenkopfs wird die Variable i wieder vergessen**|
|4 | - | die Turtle bewegt sich noch 100 vorwärts und das Programm wird beendet.|
