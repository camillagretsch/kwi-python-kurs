## 1. Was ist ein Jupyter Notebook und wie ist es aufgebaut?

Ein Jupyter Notebook hat die Endung **.ipynb**. Es ist wie ein interaktives Script, das aus Textzellen und ausführbaren Codezellen besteht. 

Jupyter bietet verschiedenste nützliche Funktionen.

---

### 1.1. Zellen

Es gibt zwei Arten von Zellen:

* **Textzellen** können Text enthalten, aber auch Links zu Webinhalten wie Filmen und Bilder.
* In **Codezellen** können Sie ausführbaren Code schreiben und direkt ausführen. 

Die Codebeispiele aus dem Script müssen Sie somit nicht abtippen! Sie können die Codezellen direkt bearbeiten und ausführen.

Ausserdem können Sie Erklärungen in Textzellen bearbeiten und ergänzen oder neue Textzellen mit eigenen Notizen hinzufügen. 

Machen Sie von der Möglichkeit Gebrauch, Ihre Notizen direkt in die Jupyter Notebooks zu machen. Damit haben Sie alles zusammen: Die Erklärungen, die Aufgaben, Ihre Lösungen und das Resultat des ausgeführten Codes.

#### 1.1.1. Zellen erstellen

Um neue Zellen zu erstellen, klicken Sie auf das `+ Code` oder `+ Markdown` im Menü oben.

Neue Zellen werden immer unterhalb der angewählten Zelle eingefügt und sind *standardmässig vom Typ «Code»*.

Angewählte Zellen lassen sich mit den Pfeilbuttons (hellblau markiert) nach oben bzw. unten verschieben.

#### 1.1.2. Zellen ausführen

Am einfachsten führen Sie Zellen mit der **Tastenkombination** [⇧]+[↩︎] (([Hochstelltaste]+[Eingabetaste]) bzw. [Shift]+[Enter]) aus.

Analog können Sie die Zelle auch durch Anklicken anwählen und mit dem Run-Knopf (►) oben im Menü ausführen.

Nachdem die Zelle ausgeführt wurde, steht zwischen den eckigen Klammern links der Zelle eine Zahl, z.B. [4].  
[\*] links der Zelle bedeutet, dass die Zelle gerade ausgeführt wird und [ ] bedeutet, dass die Zelle noch nicht ausgeführt wurde.

Beim Ausführen werden

* Textzellen (bzw. Markdownzellen) dargestellt und
* Codezellen ausgeführt.

Die Syntax von Markdown ist ziemlich einfach. Mit wenigen Elementen kommen Sie schon weit:

* **Hervorhebungen** im Text werden \*\***fett**\*\* (zwischen je *zwei* Sternen) oder \**kursiv*\* (zwischen je *einem* Stern).

* **Linien** werden mit 3 Sternen (\*\*\*) oder 3 Trennstrichen (---) markiert.

### Überschriften 
Überschriften sind jeweils durch eine oder mehrere Rauten gekennzeichnet, je weniger Rauten desto grösser der Titel. Die Anzahl der Rauten bestimmt somit die Hierarchiestufe:

# Stufe 1

Wird in der Regel nur für den Haupttitel des Jupyter Notebooks verwendet.
    
## Stufe 2
    
Ist schon etwas kleiner und würde einem Kapitel oder einem Abschnitt entsprechen (Nummerierung: 1.).
    
Und mit jeder Raute wird der Titel noch eine Stufe kleiner:
    
### Stufe 3

Ist noch etwas kleiner und würde einem Unterkapitel entsprechen oder einem Unterabschnitt (Nummerierung: 1.1.).

#### Stufe 4

Nummerierung: 1.1.1.
    
##### Stufe 5

Nummerierung: 1.1.1.1.
    
Sie könnten auch noch weitere Stufen machen, diese würden aber nicht mehr anders dargestellt.
    
### Text

Text schreiben Sie ganz normal, **fetten Text** markieren Sie entweder mit zwei `*` oder zwei `_`, kursiven Text mit jeweils einem `*` bzw. `_`. Fett und kursiv würde einer Kombination oder drei `*` bzw. `_` entsprechen.
    
* Ein Sternchen oder Underscore: *kursiv*, _kursiv_  
* Zwei Sternchen oder Underscores: **fett**, __fett__  
* Auch mi**tt**en im Wort können Sie etwas her*vor*heben.
* Auch fett-kursive Her***vor***hebungen sind möglich, zwischen `***` oder `___` oder einer Kombination aus Sternchen und Underscores: **_fettkursiv_**, __*fettkursiv*__.

### Darstellung von Code

wird


### 2.1. Befehl  `print()`

Es ist immer spannend und oft zum besseren Verständnis eines Programms hilfreich, wenn man sieht, was ein Programm macht. Dazu dient der wohl meistgenutzte Befehl `print()`.


`print()` ist ein **Befehl**, mit dem in der Konsole alles ausgegeben wird, was zwischen den Klammern steht. `print()` in Python macht am Ende immer gleich noch einen Zeilenumbruch, damit die nächste Ausgabe auf einer neuen Zeile erscheinen wird. Zwischen den Klammern kann ein Wert oder mehrere stehen.

### 2.2. Datentyp Integer und String

Damit das Programm weiss, ob mit `5` eine Zahl oder ein Text gemeint ist, werden sogenannte **Datentypen** unterschieden.

Es ist wichtig, dass das Programm unterscheiden kann, ob eine Zahl oder Text gemeint ist, damit es korrekt mit dem Wert `5` umgehen kann.

Alle ganzen Zahlen haben den Datentyp ***Integer***.

Alles, was **Text** ist, hat den Datentyp ***String*** (für «Zeichenkette»). Damit das Programm weiss, dass es sich bei einem Wert um einen String handelt, werden Strings durch einfache oder doppelte Hochkommas gekennzeichnet:

```python

"5", "Hallo", "1<4", "1.1.2023"
```

sind alles Strings, genauso wie:

```python
'-7', 'KWI', '2*4', 'Ich bin ein String!'
```

Führen Sie die folgenden vier Codezellen aus und beobachten Sie, was mit den beiden Datentypen geschieht.

````python
print(5)
````

````python
print("5")
````

````python
print(5+2)
````

````python
print("5+2")
````
